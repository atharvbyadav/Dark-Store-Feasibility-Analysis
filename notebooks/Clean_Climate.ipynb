```python
import pandas as pd
import numpy as np

# Load raw data
df_raw = pd.read_csv('Pune_Raw_Climate_Data.csv')
```

```python
# Fill missing temperature values with their column median
df_raw['Avg High Temp (°C)'].fillna(df_raw['Avg High Temp (°C)'].median(), inplace=True)
df_raw['Avg Low Temp (°C)'].fillna(df_raw['Avg Low Temp (°C)'].median(), inplace=True)
```

```python
# Fill missing rainfall values with 0 (assuming no rain if data is missing)
df_raw['Avg Rainfall (mm)'].fillna(0, inplace=True)
```

```python
# Fill missing delivery impact scale with mode (most frequent value)
df_raw['Delivery Impact Scale (1-5)'].fillna(df_raw['Delivery Impact Scale (1-5)'].mode()[0], inplace=True)
df_raw.to_csv('Pune_Cleaned_Climate_Data.csv', index=False)
print('Cleaning complete. Cleaned CSV file saved.')
```
