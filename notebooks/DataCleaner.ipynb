```python
import pandas as pd
import numpy as np

# Load raw data
raw_population = pd.read_csv('Pune_Raw_Population_Data.csv')
raw_online_activity = pd.read_csv('Pune_Raw_Online_Activity_Data.csv')
```

```python
# Fill missing population data with median population
median_population = raw_population['Estimated Population (2024)'].median()
raw_population['Estimated Population (2024)'].fillna(median_population, inplace=True)
```

```python
# Assume internet penetration based on city average (75%) if missing
raw_online_activity['Internet Penetration Rate (%)'].fillna(75, inplace=True)
```

```python
# Estimate online shoppers based on penetration rate
raw_online_activity['Estimated Online Shoppers'] = (raw_population['Estimated Population (2024)'] * \
                  (raw_online_activity['Internet Penetration Rate (%)'] / 100)).astype(int)
```

```python
# Fill missing e-commerce activity index with average value
raw_online_activity['E-Commerce Activity Index (1-10)'].fillna(7, inplace=True)
```

```python
# Estimate predicted online order volume
raw_online_activity['Predicted Online Order Volume (Monthly)'] = (raw_online_activity['Estimated Online Shoppers'] * \
(raw_online_activity['E-Commerce Activity Index (1-10)'] / 10)).astype(int)
```

```python
# Save cleaned data
raw_population.to_csv('Pune_Cleaned_Population_Data.csv', index=False)
raw_online_activity.to_csv('Pune_Cleaned_Online_Activity_Data.csv', index=False)
print('Cleaning complete. Cleaned CSV files saved.')
```
